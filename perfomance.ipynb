```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('detailed_performance_results.csv')
```

```python
for function in data['Function'].unique():
    plt.figure(figsize=(10, 5))
    subset = data[data['Function'] == function]
    sns.lineplot(x='Size', y='Time', data=subset, marker='o')
    plt.title(f'Performance of {function}')
    plt.xlabel('Array Size')
    plt.ylabel('Time (ms)')
    plt.grid(True)
    plt.savefig(f'{function}_performance.png')
    plt.close()
```
